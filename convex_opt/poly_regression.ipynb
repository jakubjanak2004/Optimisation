```python
import numpy as np
import matplotlib.pyplot as plt
from polynomial_regression import PolynomialRegression
```

```python
# Creating an synthetic data to try polynomial regression on
num_of_samples = 100
std_dev = 250
x = np.linspace(-10, 10, num_of_samples)
y = x ** 3 + np.random.randn(num_of_samples) * std_dev
```

```python
# Calculating Polynomial regression model and plotting it with the training data
pr = PolynomialRegression(n=3)
pr.train(
    x=x,
    y=y
)

pr.plot_trained_model()
```
